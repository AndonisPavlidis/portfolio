```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

sns.set_palette("colorblind", n_colors=10)

palette = sns.color_palette(palette="colorblind", n_colors=10)
sns.palplot(palette)
```

## Load Data

https://github.com/mwaskom/seaborn-data/blob/master/tips.csv

```python
tips = sns.load_dataset("tips")
```

## Histograms

```python
fig = plt.figure(figsize=(8, 5))
ax = plt.subplot(1, 1, 1)

sns.distplot(
    tips["total_bill"], 
    kde=False, 
    hist_kws=dict(alpha=1), 
    ax=ax
)

ax.set_title("Histogram of Total Bill", fontsize=16)
ax.set_xlabel("Amount ($)", fontsize=12)

plt.show()
```

```python
fig = plt.figure(figsize=(8, 5))
ax = plt.subplot(1, 1, 1)
sns.distplot(
    tips["total_bill"], 
    kde=False, 
    label="total_bill", 
    hist_kws=dict(alpha=1), 
    ax=ax
)

sns.distplot(
    tips["tip"], 
    kde=False, 
    label="tip", 
    color="#fc6443", 
    hist_kws=dict(alpha=1), 
    ax=ax
)


ax.set_title("Histogram of Both Tip Size and Total Bill", fontsize=16)
ax.set_xlabel("Amount ($)", fontsize=12)
ax.legend()

plt.show()
```

## Boxplots

```python
fig = plt.figure(figsize=(20, 5))

ax1 = plt.subplot(1, 2, 1)
sns.boxplot(tips["total_bill"], ax=ax1)
ax1.set_title("Boxplot of Total Bill", fontsize=16)
ax1.set_xlabel("Amount ($)", fontsize=12)

ax2 = plt.subplot(1, 2, 2)
sns.boxplot(tips["tip"], ax=ax2)
ax2.set_title("Boxplot of Tip", fontsize=16)
ax2.set_xlabel("Amount ($)", fontsize=12)

plt.show()
```

## Boxplots and Historgrams

```python
fig = plt.figure(figsize=(20,10))

grid = plt.GridSpec(2, 2, wspace=0.1, hspace=0.3)

ax1 = plt.subplot(grid[:, 0])
sns.boxplot(
    x = tips["tip"], 
    y = tips["time"], 
    palette=[palette[0], palette[2]],
    saturation=1,    
    ax=ax1
)
ax1.set_title("Tip by Time of day", fontsize=16)
ax1.set_xlabel("Amount ($)", fontsize=12)
ax1.set_ylabel("Time", fontsize=12)

ax2 = plt.subplot(grid[0, 1])
sns.distplot(
    tips.loc[tips["time"]=="Lunch", "tip"], 
    kde=False,
    hist_kws=dict(alpha=1), 
    ax=ax2
)
ax2.set_title("Lunch", fontsize=16)
ax2.set_xlabel("Amount ($)", fontsize=12)
ax2.set_xlim(0,12)

ax3 = plt.subplot(grid[1, 1])
sns.distplot(
    tips.loc[tips["time"]=="Dinner", "tip"], 
    kde=False, 
    color=palette[2],
    hist_kws=dict(alpha=1), 
    ax=ax3
)
ax3.set_title("Dinner", fontsize=16)
ax3.set_xlabel("Amount ($)", fontsize=12)
ax3.set_xlim(0,12)


plt.show()
```
