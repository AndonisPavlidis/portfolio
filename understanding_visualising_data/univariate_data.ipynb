```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

### Load Data

https://github.com/mwaskom/seaborn-data/blob/master/iris.csv

```python
tips = sns.load_dataset("tips")
```

### Tables

```python
tips.head()
```

### Describe

```python
tips.describe()
```

### Histograms

```python
fig = plt.figure(figsize=(8, 5))
ax = plt.subplot(1, 1, 1)

ax = sns.distplot(tips["total_bill"], kde=False)

ax.set_title("Histogram of Total Bill", fontsize=16)
ax.set_xlabel("Amount ($)")

plt.show()
```

```python
fig = plt.figure(figsize=(8, 5))
ax = plt.subplot(1, 1, 1)
sns.distplot(tips["tip"], kde=False, label="tip", color="#fc6443")
sns.distplot(tips["total_bill"], kde=False, label="total_bill")

ax.set_title("Histogram of Both Tip Size and Total Bill", fontsize=16)
ax.set_xlabel("Amount ($)")
ax.legend()

plt.show()
```

### Boxplots

```python
fig = plt.figure(figsize=(20, 5))
ax1 = plt.subplot(1, 2, 1)

ax1 = sns.boxplot(tips["total_bill"])
ax1.set_title("Boxplot of Total Bill", fontsize=16)
ax1.set_xlabel("Amount ($)")

ax2 = plt.subplot(1, 2, 2)
ax2 = sns.boxplot(tips["tip"])
ax2.set_title("Boxplot of Tip", fontsize=16)
ax2.set_xlabel("Amount ($)")

plt.show()
```

### Boxplots and Historgrams

```python
fig = plt.figure(figsize=(20,10))

grid = plt.GridSpec(2, 2, wspace=0.1, hspace=0.3)
ax1 = plt.subplot(grid[:, 0])

ax1 = sns.boxplot(x = tips["tip"], y = tips["time"])
ax1.set_title("Boxplot of Tip by Time of day", fontsize=16)
ax1.set_xlabel("Amount ($)")

ax2 = plt.subplot(grid[0, 1])

ax2 = sns.distplot(tips.loc[tips["time"]=="Lunch", "tip"])
ax2.set_title("Histogram of Tip during Lunch", fontsize=16)
ax2.set_xlabel("Amount ($)")
ax2.set_xlim(0,12)

ax3 = plt.subplot(grid[1, 1])

ax3 = sns.distplot(tips.loc[tips["time"]=="Dinner", "tip"])
ax3.set_title("Histogram of Tip during Dinner", fontsize=16)
ax3.set_xlabel("Amount ($)")
ax3.set_xlim(0,12)


plt.show()
```
