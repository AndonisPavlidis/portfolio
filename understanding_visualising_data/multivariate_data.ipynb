```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

## Load data

https://github.com/mwaskom/seaborn-data/blob/master/iris.csv

```python
iris = sns.load_dataset("iris")
```

## Tables

```python
iris.head()
```

## Describe

```python
iris.describe()
```

## Scatter

```python
plt.figure(figsize=(8,8))

ax = plt.subplot(1,1,1)

sns.scatterplot(
    x="sepal_length", 
    y="petal_length", 
    data=iris, 
    hue="species", 
    ax=ax)

ax.set_title("Sepal Length vs Petal Length", fontsize=16)
ax.set_xlabel("Sepal Length", fontsize=12)
ax.set_ylabel("petal Length", fontsize=12)
ax.legend(fontsize=14)

plt.show()
```

##Barcharts

```python
plt.figure(figsize=(10,8))

ax = plt.subplot(1,1,1)

sns.barplot(
    x="species", 
    y="sepal_length", 
    data=iris, 
    color="grey", 
    saturation=1,
    ax=ax
)

ax.set_title("Sepal Length", fontsize=16)
ax.set_xlabel("")
ax.set_ylabel("Sepal Length", fontsize=14)
ax.tick_params(axis='x', which='major', labelsize=14)

plt.show()
```
