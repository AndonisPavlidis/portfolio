## k-Nearest Neighbours

### Show me thy neighbour and I will tell you who you are

A simple predictive model that makes no *mathematical* assumptions and does not require any heavy machinery.

It requires:
* Some notion of distance

* An asummption that points that are close to one another are similar

It neglets a lot of information, as the the prediction depends on a handful of neighbors, so will not allow you to understand the drivers of whatever phenomenon you're looking at.

### Build from Scratch: Predict a vote

```python
from typing import List
from collections import Counter

def raw_majority_vote(
    votes: List[str]
) -> str:
    vote_counts = Counter(votes)
    winner, _ = vote_counts.most_common(1)[0]
        
    return winner

votes = ['trump', 'anyone_else', 'anyone_else', 'anyone_else']
winner =  raw_majority_vote(votes)

print(f"Winner is: {winner}")
```

**What happens when we have a tie:**  
1) Pick one at random  
2) Weight the voters by distance and pick the weighted winner  
3) Reduce k until we find a unique winner  

We will implement (3)

```python
def majority_vote(
    votes: List[str]
):
    """Assumes labels are ordered from nearest to farthest."""
    draw = True
    
    while draw:
        vote_counts = Counter(votes)
        winner, winner_count = vote_counts.most_common(1)[0]

        num_winners = len(
            [count for count in vote_counts.values() if count==winner_count]
        )
    
        if num_winners != 1:
            votes.pop()
        else:
            draw = False
            
    return winner

votes = ['trump', 'someone_else', 'anyone_else', 'someone_else', 'anyone_else']
print(majority_vote(votes))
```
